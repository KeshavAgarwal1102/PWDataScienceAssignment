```python
# Here are the key features of lists in Python:

#  1. Ordered
  #  - Lists maintain the order of elements. The order in which items are inserted is preserved when accessing them.

     fruits = ["apple", "banana", "cherry"]
     print(fruits[0])  # Output: apple


# 2. Mutable
  #  - Lists are mutable, meaning their elements can be changed after the list is created.

     fruits = ["apple", "banana", "cherry"]
     fruits[1] = "orange"
     print(fruits)  # Output: ['apple', 'orange', 'cherry']


# 3. Can Contain Different Data Types
  #  Lists can hold elements of different data types (e.g., integers, strings, or even other lists).

     mixed_list = [1, "apple", 3.14, [1, 2, 3]]
     print(mixed_list)  # Output: [1, 'apple', 3.14, [1, 2, 3]]


# 4. Indexed
  #   Lists are indexed, meaning each element has a unique index, starting from `0` for the first element.

     fruits = ["apple", "banana", "cherry"]
     print(fruits[1])  # Output: banana


 # 5. Supports Slicing
  #  Lists support slicing, allowing you to access a range of elements using a start and end index.

     fruits = ["apple", "banana", "cherry", "date"]
     print(fruits[1:3])  # Output: ['banana', 'cherry']


### 6. Dynamic Size
  #   Lists are dynamic, meaning their size can grow or shrink as items are added or removed.

     fruits = ["apple", "banana"]
     fruits.append("cherry")  # Adds "cherry" to the list
     print(fruits)  # Output: ['apple', 'banana', 'cherry']


# 7. Allow Duplicates
    # Lists can contain duplicate elements. The same item can appear multiple times.

     fruits = ["apple", "banana", "apple"]
     print(fruits)  # Output: ['apple', 'banana', 'apple']


# 8. Supports Iteration
    # Lists are iterable, meaning you can loop over the elements using a `for` loop.

     fruits = ["apple", "banana", "cherry"]
     for fruit in fruits:
         print(fruit)
     # Output: apple
     #         banana
     #         cherry


# 9. Can Contain Nested Lists
  #  Lists can contain other lists, making it possible to create multi-dimensional lists (e.g., matrices).

     nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
     print(nested_list[0])  # Output: [1, 2, 3]


# 10. Built-in Methods
  #  Lists come with several built-in methods like `append()`, `remove()`, `insert()`, `sort()`, and more to modify or access the elements.

     fruits = ["apple", "banana", "cherry"]
     fruits.append("date")  # Adds "date"
     fruits.remove("banana")  # Removes "banana"
     fruits.sort()  # Sorts the list alphabetically
     print(fruits)  # Output: ['apple', 'cherry', 'date']

```
