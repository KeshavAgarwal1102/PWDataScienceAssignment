```python
# In Python, you can check if a file exists before attempting to read it using the os.path.exists() function from the os module or pathlib.Path.exists() from the pathlib module. Here's how you can do it:

# Using os.path.exists():

import os

def read_file_if_exists(filename):
    if os.path.exists(filename):  # Check if the file exists
        try:
            with open(filename, 'r') as file:
                content = file.read()
                print("File content:")
                print(content)
        except Exception as e:
            print(f"An error occurred while reading the file: {e}")
    else:
        print(f"Error: The file '{filename}' does not exist.")

# Example usage
filename = input("Enter the file name: ")
read_file_if_exists(filename)



# Using pathlib.Path.exists():

from pathlib import Path

def read_file_if_exists(filename):
    file_path = Path(filename)
    if file_path.exists():  # Check if the file exists
        try:
            with open(filename, 'r') as file:
                content = file.read()
                print("File content:")
                print(content)
        except Exception as e:
            print(f"An error occurred while reading the file: {e}")
    else:
        print(f"Error: The file '{filename}' does not exist.")

# # Example usage
# filename = input("Enter the file name: ")
# read_file_if_exists(filename)
# Explanation:
# Using os.path.exists():

# os.path.exists(filename) checks if the file exists at the given path. It returns True if the file exists, and False if it does not.
# Using pathlib.Path.exists():

# Path(filename).exists() checks if the file exists using the pathlib module, which is more modern and object-oriented compared to os.path.
# File Handling:

# If the file exists, the program opens it, reads its contents, and prints them.
# If the file doesn't exist, an appropriate error message is displayed.
# You can also handle exceptions like FileNotFoundError or others, but checking with exists() is a good first step to prevent errors.
```
