```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 8, 7]

# Create a scatter plot
plt.scatter(x, y, color='blue', marker='o', label='Data Points')

# Add titles and labels
plt.title('Scatter Plot Example')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Add a legend
plt.legend()

# Show the plot
plt.show()

# Explanation:
# Data Preparation:

# Define x and y lists containing the coordinates of the points.
# Scatter Plot Creation:

# Use plt.scatter() to create a scatter plot.
# Customize the color, marker type, and add a label.
# Adding Titles and Labels:

# Use plt.title(), plt.xlabel(), and plt.ylabel() to add meaningful text to the plot.
# Adding a Legend:

# Use plt.legend() to display the label for the scatter points.
# Displaying the Plot:

# Use plt.show() to render the plot.
```
